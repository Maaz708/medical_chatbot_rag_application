```python
from langchain_core.prompts import PromptTemplate
```

```python
prompt_template1 = PromptTemplate(input_variables=["domain"],
                                 template = """
                                 I want to known free resources
                                 to learn about {domain}. Give me youtube and free learnig 
                                 resource in a structured way.

                                 Give me a json output.
                                 Do not put markdown tags and any other tags.
                                 Only give json response.
                                 """
                                )
```

```python
input_prompt = prompt_template1.format(domain="AI engineer")
```

```python
print(input_prompt)
```

```python
prompt_template2 = PromptTemplate(input_variables=["company"],
                                 template = """
                                 You are a JSON API
                                 give only valid JSON
                                 
                                 Provide me free courses of
                                 Company: {company} with free certificates.
                                
                                 Output format:
                                 {{
                                 Courses:[
                                 {{
                                 "name":"",
                                 "platform":"",
                                 "certificate": True,
                                 "url":"",
                                 }}
                                 ]
                                 }}
                                 return only json.
                                 """
                                )
```

```python
input_prompt2 = prompt_template2.format(company="Google")
```

```python
print(input_prompt2)
```

```python
chain1 = prompt_template2 | llm
response1 = chain1.invoke("Microsoft")
```
